```
import pyspark.sql.functions as f#Read in data

df_got_info = spark.sql("SELECT * FROM got_csv_laura2_csv")
df_got_sub = spark.sql("SELECT * FROM subtitle_list_laura_csv")

#df_got_sub.show()

df_got_sub.show()
df_got_info.show()
```

```
# implement word count for each line

import pyspark.sql.functions as psf

df = df_got_sub.withColumn('wordCount', psf.size(psf.split(psf.col('value'), ' ')))
df.show()
df.count()
```

```
#Word Filter (OR)

words_list = ['wolf', 'Winterfell', 'north']
df.filter(psf.col('value').rlike('(^|\s)(' + '|'.join(words_list) + ')(\s|$)'))
df_selected_episodes = df.filter(psf.col('value').rlike('(^|\s)(' + '|'.join(words_list) + ')(\s|$)'))

df_selected_episodes.show()
df_selected_episodes.count()
```

```
#Liste (df) mit unterschiedlichen Zeileneinträgen:

df_subset = df_selected_episodes.select('episode_nr').distinct()
                                                                          
df_subset.show()
df_subset.count()

```

```
#subsetting dataframe

df.filter(df.episode_nr == 'S01E05').count()
```

```
from pyspark.sql.functions import udf, col, lower, regexp_replace, explode, split
from pyspark.ml.feature import Tokenizer, StopWordsRemover

#2 Anwendungsfälle:
#df_selected: eine bestimmte Episode (siege Filter oben)
#df_selected_episodes: alle Episoden mit den gesuchten Begriffen aus der Wortliste oben

#df_selected = df.filter(df.episode_nr == 'S01E05')
df_selected = df_selected_episodes

words_list_clean = df_selected.withColumn('value', regexp_replace('value', '[\\,,\\.,\\?,\\-,\\!]', ''))

words_list_clean = words_list_clean.withColumn('value', lower(col('value')))

single_words = words_list_clean.select(explode(split(col("value"), "\s+")).alias("value"))
#single_words.show()

tokenizer = Tokenizer(inputCol='value', outputCol='words_token')
tokenized_list = tokenizer.transform(single_words)
#tokenized_list.show()

#remove stop-words and filter empty lists (stop words)
remover = StopWordsRemover(inputCol='words_token', outputCol='cleaned')
single_words_cleaned = remover.transform(tokenized_list).filter(psf.size('cleaned') > 0)

single_words_cleaned.show()
```

```
#Worte zählen, gruppieren, sortieren und ausgeben.
wordCountsDF = (single_words_cleaned
                .groupBy('value').count())
wordCountsDF.sort(col("count").desc()).show()
```

```
#Loop für alle Episoden:

df_subset.show()
result = []
for i in df_subset:
  wordCountsDF = (single_words_cleaned.groupBy('value').count())
  wordCountsDF.sort(col("count").desc())
  wordCountsDF.show()
  result.append((i,wordCountsDF))

print(result)
```
